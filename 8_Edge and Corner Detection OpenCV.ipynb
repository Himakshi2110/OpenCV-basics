# <center> Edge detection using OpenCV

## Different types of edge detection in openCV 

* Canny Edge detection
* Sobel Edge detection
* Laplacian Edge detection

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

   
def image_plot_multiple(title,image): ## display function
    L=len(title)
    ncols=min(L,3)
    nrows=int(np.ceil(L/ncols))
    plt.subplots(nrows,ncols,figsize=(7*ncols,nrows*5))
    for i,item in enumerate(title):
        plt.subplot(nrows,ncols,i+1)
        plt.imshow(cv2.cvtColor(image[i],cv2.COLOR_BGR2RGB))
        plt.title(title[i])
```

```python
lion_path=os.path.join('images','lion.jpeg') 
horse_image=os.path.join('images','Dreaming_about_a_horse.jpg')
lion=cv2.imread(lion_path,0)
horse=cv2.imread(horse_image,0)
```

# Sobel Edge Detection


```python
sobelx=cv2.Sobel(horse,cv2.CV_8U,0,1,ksize=7)
sobely=cv2.Sobel(horse,cv2.CV_8U,1,0,ksize=7)
Sobel_edges=cv2.bitwise_or(sobelx,sobely)
plt.imshow(lion,cmap='gray')
plt.title('Original')
image_plot_multiple(['Horizontal edges','Vertical Edges','Both'],[sobelx, sobely, Sobel_edges])
```

```python
image=cv2.imread(os.path.join('images','perspective.jpeg'),0)
sobelx=cv2.Sobel(image,cv2.CV_8U,0,1,ksize=7)
sobely=cv2.Sobel(image,cv2.CV_8U,1,0,ksize=7)
Sobel_edges=cv2.bitwise_or(sobelx,sobely)
plt.imshow(image,cmap='gray')
plt.title('Original')
image_plot_multiple(['Horizontal edges','Vertical Edges','Both'],[sobelx, sobely, Sobel_edges])
```

# Laplacian Edge Detection

```python
image=cv2.imread(os.path.join('images','Dreaming_about_a_horse.jpg'),0)
Laplacian_edges=cv2.Laplacian(image,cv2.CV_8U,ksize=7)
image_plot_multiple(['Original','Laplacian Edges'],[image, Laplacian_edges])
```

```python
image=cv2.imread(os.path.join('images','perspective.jpeg'),0)
Laplacian_edges=cv2.Laplacian(image,cv2.CV_8U,ksize=7)
image_plot_multiple(['Original','Laplacian Edges'],[image, Laplacian_edges])
```

```python
image=cv2.imread(os.path.join('images','lion.jpeg'),0)
Laplacian_edges=cv2.Laplacian(image,cv2.CV_8U,ksize=7)
image_plot_multiple(['Original','Laplacian Edges'],[image, Laplacian_edges])
```

## Canny Edge Detection

```python
image=cv2.imread(os.path.join('images','perspective.jpeg'),0)
Canny_edges=cv2.Canny(image,100,255)
image_plot_multiple(['Original','Canny Edges'],[image, Canny_edges])
```

```python
image=cv2.imread(os.path.join('images','Dreaming_about_a_horse.jpg'),0)
Canny_edges=cv2.Canny(image,200,255)
image_plot_multiple(['Original','Canny Edges'],[image, Canny_edges])
```

```python
image=cv2.imread(os.path.join('images','lion.jpeg'),0)
Canny_edges=cv2.Canny(image,150,250)
image_plot_multiple(['Original','Canny Edges'],[image, Canny_edges])
```

# Corner Detection

```python
import cv2
import numpy as np
image=cv2.imread(os.path.join('images','perspective.jpeg'))
image1=np.copy(image)
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
gray = np.float32(gray)
dst = cv2.cornerHarris(gray,blockSize=2,ksize=3,k=0.04)
dst = cv2.dilate(dst,None)
image1[dst>0.2*dst.max()]=[0,0,255]
image_plot_multiple(['Original','Corners'],[image, image1])
```
